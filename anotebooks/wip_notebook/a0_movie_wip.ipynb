# GLOBAL Variables

## CLOUD_THRESHOLD 

## FILL_THRESHOLD

- CLOUD_THRESHOLD = 10 - only 10 percent or less cloud - clear images in other words
- FILL_THRESHOLD = 99 - essentially all fill levels

```python
CLOUD_THRESHOLD = 10 ### - only 10 percent or less cloud - clear images in other words
FILL_THRESHOLD = 99 ### - essentially all fill levels
```

```python
import datacube
import time

from datetime import datetime

dc = datacube.Datacube()

selected_product = 'landsat_8_USARD'

date_range = (
        datetime(1984,1,1),
        datetime(2018,12,31))
dss = dc.find_datasets(product=selected_product, time=date_range, measurements=['red',])

clear_datasets = []
for item in dss:
    if float(item.metadata_doc['cloud_cover']) < CLOUD_THRESHOLD:
        # print(item.metadata_doc['cloud_cover'])
        if float(item.metadata_doc['fill']) < FILL_THRESHOLD:
            # print(item.metadata_doc['fill'])
            clear_datasets.append(item)
            
print("Number of Pretty Scenes", len(clear_datasets))
for citem in clear_datasets:
    print(citem.uris)

    


```

```python
Start = time.time()
```

```python
ds_l8 = dc.load(product=selected_product, datasets=clear_datasets, measurements=['red','green','blue'], x = (-122.65,-122.75), y =(45.59,45.65),output_crs = 'epsg:5072', resolution = (-30,30))
```

```python
End = time.time()

print ("Load time took seconds = ", End-Start)
```

```python
from rasterio.plot import show_hist
from matplotlib import pyplot
%matplotlib inline
from noteLib import *
def plot_me(ds, time, count, sensor='L8'):
    figsize = [8,8]
    
    img_toshow = threeBandImage(ds,bands = ['red', 'green', 'blue'], time = time)
    fig = pyplot.figure(figsize = figsize)
    pyplot.imshow(img_toshow)
    ax = pyplot.gca()
    title_string = "--> IMG " + str(count) + ' ' + sensor
    figName = title_string + '.png'
    title_string = str(ds.time[time].values).split('T')[0] + title_string
    time_string = str(ds.time[time].values).split('T')[0]
    figName = time_string + '--' + sensor + '.png'
    ax.set_title(title_string, fontweight = 'bold', fontsize = 16)
    ax.set_xticklabels(ds.x.values)
    ax.set_yticklabels(ds.y.values)
    ax.set_xlabel('Easting', fontweight = 'bold')
    ax.set_ylabel('Northing', fontweight = 'bold')
    print(figName)
    pyplot.savefig(figName)
    # if you remove the close below the images will display - uses lots of memory though
    pyplot.close(fig)
    return True
```

```python
ntime = len(ds_l8.time)
cnt = 0
for t in range (ntime):
    # print("Hi",cnt)
    if ds_l8.red[t,0,0] > 0:
        cnt = cnt + 1
        print("CNT", cnt)
        plot_me(ds_l8, t, cnt, 'L8')
```
