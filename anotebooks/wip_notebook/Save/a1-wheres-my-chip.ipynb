```python
def return_tile_corners(h,v):

    conus_ulX = -2565585 # - longitude
    conus_ulY = 3314805

    span = 5000 * 30

    c_ulX = conus_ulX + (h * span)
    c_ulY = conus_ulY - (v * span)

    c_lrX = c_ulX + span
    c_lrY = c_ulY - span

    return (c_ulX, c_ulY, c_lrX, c_lrY)
```
