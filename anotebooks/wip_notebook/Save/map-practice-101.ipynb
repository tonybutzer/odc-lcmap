```python
import datacube
import time

from datetime import datetime


dc = datacube.Datacube(app = 'my_app', config = '../datacube.conf')
selected_product = 'landsat_8_USARD'

date_range = (
        datetime(2015,6,1),
        datetime(2015,7,31))
dss = dc.find_datasets(product=selected_product, time=date_range, measurements=['red',])

clear_datasets = []
for item in dss:
    
    if float(item.metadata_doc['cloud_cover']) < 20.0:
        if float(item.metadata_doc['fill']) < 30.0:
            clear_datasets.append(item)
            
print("Number of Pretty Scenes", len(clear_datasets))
```

```python
citem = clear_datasets[0]
json_in = citem.metadata_doc



longitude_extents = ((float)(json_in['extent']['coord']['ll']['lon']), (float)(json_in['extent']['coord']['ur']['lon']))

print("longitude_extents",longitude_extents)

latitude_extents = ((float)(json_in['extent']['coord']['ll']['lat']), (float)(json_in['extent']['coord']['ur']['lat']))

print("latitude_extents",latitude_extents)
```

```python
import itertools    
import math
import numpy as np  

def _degree_to_zoom_level(l1, l2, margin = 0.0):
    
    degree = abs(l1 - l2) * (1 + margin)
    zoom_level_int = 0
    if degree != 0:
        zoom_level_float = math.log(360/degree)/math.log(2)
        zoom_level_int = int(zoom_level_float)
    else:
        zoom_level_int = 18
    return zoom_level_int
```

```python
from pyproj import Proj
def _calc_circle_list(lat,lon):
    p = Proj(init='epsg:5072') # EPSG code AEA

    l=citem.bounds.left
    t=citem.bounds.top
    r=citem.bounds.right
    b=citem.bounds.bottom
    
    y_d = t-b
    y2 = y_d /30
    y3 = y2/100
    print("y_d = ",y_d)
    print(y2,y3)
    circle_list =[]
    for x in range(1,50):
        y = -x
        print(x,y)
        # aea_ul = p(lon[0],lat[1])
        #print(aea_ul)
        #cll = p(aea_ul[0], aea_ul[1], inverse=True)
        #print(cll)
        aea_ul = (l+(x*100*30), t+(y*100*30))
        print(aea_ul)
        cll = p(aea_ul[0], aea_ul[1], inverse=True)
        print(cll)
        clatlon = (cll[1],cll[0])
        circle_list.append(clatlon)
    print(l,t)
    print(r,b)
    return circle_list
```

```python
circ_list = _calc_circle_list(latitude_extents, longitude_extents)
```

```python
import folium
import numpy as np
the_cll = circ_list[0]
location = [latitude_extents[0], longitude_extents[0]]
latitude = latitude_extents
longitude = longitude_extents
center = [np.mean(latitude), np.mean(longitude)]

margin = -0.5
zoom_bias = 0
    
lat_zoom_level = _degree_to_zoom_level(margin = margin, *latitude ) + zoom_bias
lon_zoom_level = _degree_to_zoom_level(margin = margin, *longitude) + zoom_bias
zoom_level = min(lat_zoom_level, lon_zoom_level) 

sg_map = folium.Map(location=center, zoom_start=zoom_level,
                    tiles=" http://mt1.google.com/vt/lyrs=y&z={z}&x={x}&y={y}",
                    attr="Google via tony"
                   )

line_segments = [(latitude[0],longitude[0]),
                     (latitude[0],longitude[1]),
                     (latitude[1],longitude[1]),
                     (latitude[1],longitude[0]),
                     (latitude[0],longitude[0])
                    ]

    
sg_map.add_child(folium.features.PolyLine(locations=line_segments,color='red',opacity=0.8))
# for record in records:
  #  lat, lng = svy.computeLatLon(record.shape.points[0][1], record.shape.points[0][0])
clat,clng = the_cll
print(clat,clng)


for i in range(0,len(circ_list)):
    clat,clng = circ_list[i]
    print(clat,clng)
    folium.Circle(
        location=[clat,clng],
        radius=50*30,
        popup='i',
        color='#3186cc',
        fill=True,
        fill_color='#3186cc'
).add_to(sg_map)

folium.Marker([clat,clng], popup='<i>Mt. WTF Meadows</i>').add_to(sg_map)

folium.Marker([45.6437,-122.7474], popup='<i>Tip of Hayden Island</i>').add_to(sg_map)


# sg_map.save("tony.html")
sg_map
```
