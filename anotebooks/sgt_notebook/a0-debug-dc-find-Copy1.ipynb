```python
def display_bound(ds):
    
    myds = ds[0]

    ulX = myds.bounds.left
    ulY = myds.bounds.top

    lrX = myds.bounds.right
    lrY = myds.bounds.bottom
    m = aea_base_map(ulX,ulY,lrX,lrY)
    
    for i in range(0,len(ds)):
        myds = ds[i]

        ulX = myds.bounds.left
        ulY = myds.bounds.top

        lrX = myds.bounds.right
        lrY = myds.bounds.bottom
    
        aea_display_map(m,ulX,ulY,lrX,lrY)
        
    return(m)
        
        
```
