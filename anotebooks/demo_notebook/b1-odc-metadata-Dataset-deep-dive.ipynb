## Where are the clouds in this chip?

```python
# remember the .43 cloud cover value - where are they?
```

```python
def name_that_qa(pixel):

    qa_dict = { 322: 'Clear terrain, low-confidence cloud, low-confidence cirrus',
                324: 'Water, low-confidence cloud, low-confidence cirrus',
                328: 'Cloud shadow, low-confidence cloud, low-confidence cirrus',
                352: 'Cloud, low-confidence cloud, low-confidence cirrus',
                416: 'Cloud, medium-confidence cloud, low-confidence cirrus',
                480: 'Cloud, high-confidence cloud, low-confidence cirrus',
               
              }
    
    return qa_dict[pixel]
```

```python
qa_cnt = {}
ary = dc_xarray['pixel_qa'][0].values
for x in range(0,100):
    for y in range (0,100):
        if ary[x,y] in qa_cnt.keys():
            qa_cnt[ary[x,y]] = qa_cnt[ary[x,y]] + 1
        else:
            qa_cnt[ary[x,y]] = 1
    print
    
for i in sorted(qa_cnt):
    print(i, qa_cnt[i], name_that_qa(i))
    
print ('__' * 50)
num_pixels = 100 * 100
print(num_pixels,"Total Pixels")
for i in sorted(qa_cnt):
    pct = qa_cnt[i]/num_pixels * 100
    print(i, "%02d" % pct, '% -',name_that_qa(i))
print ('__' * 50)
```
