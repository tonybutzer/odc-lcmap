```python
import pickle

pickle_name = "h{}v{}ch{}cv{}_ccd_model_chip_array.pickle".format(h,v,ch,cv)

print(pickle_name)
pickle_in = open(pickle_name, "rb")
mo_2D_dict_array = pickle.load(pickle_in)
pickle_in.close()
```

```python
import matplotlib
from matplotlib import pyplot

# this could be in the nb_image.py file - plot_ccd_prob
def color_plot(ary):
    
    w_cmap = matplotlib.colors.ListedColormap(['#ffcc66','#ffffcc' , '#ccff66' , '#2e2eb8', '#0000ee' , '#0000ee', '#0000ee'])
    w_bounds = [0, 0.1, 0.25, 0.35, 0.5, 0.8, 1]
    w_norm = matplotlib.colors.BoundaryNorm(w_bounds,w_cmap.N)


    print(ary.shape)

    pyplot.imshow(ary, cmap=w_cmap, norm = w_norm)
    pyplot.colorbar()
    pyplot.show()
```

```python
import numpy
width = 100
height = 100

for x in range(width):
    for y in range(height):
        read_model = mo_2D_dict_array[x][y]
        if (read_model['water_prob'] > 0):
            print(x,y,read_model['water_prob'])


water_array = numpy.zeros(shape = (width,height))

for x in range(0,width):
    for y in range(0,height):
        water_array[x][y] = mo_2D_dict_array[x][y]['water_prob']
    

```

```python
color_plot(ary=water_array)
```
